```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data/IT_vacancies_full.csv' )
df
```

```python
df = df[['Name' , 'Description' , 'Experience' , 'Schedule']]
```

```python
df['Description']  = df['Description'] + ' ' + df['Experience'] + ' ' + df['Schedule']
```

```python
df.drop(['Experience' , 'Schedule' ] , axis = 1 , inplace = True)
```

```python
df['Description'][0]
```

```python
df
```

```python
train_df, val_df = train_test_split(df, test_size=0.5, random_state=42)
```

```python
train_df.to_pickle('data/all_data.pkl')
```
